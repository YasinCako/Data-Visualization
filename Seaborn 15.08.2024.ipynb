```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe().T
```

## Scatter Plot

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
plt.show()
```

```python
sns.regplot(data = tips, y = "tip", x = "total_bill");
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex")
plt.show()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker")
plt.show() 
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "day")
plt.show() 
```

## Style

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker", style = "sex")
plt.show() 
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker", style = "day")
plt.show() 
```

```python
sns.scatterplot(x = tips["total_bill"], y = tips["tip"], hue = tips.smoker, style = tips.day)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10 ,5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, style = "sex", ax = ax[0])
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker", ax = ax[1]);
```

```python

plt.figure(figsize = (10, 5))

sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "plasma")

plt.show()
```

```python
plt.figure(figsize = (10, 5))

sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "plasma", alpha = 0.5);
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips,
                hue = "sex",
                palette = "plasma",
                alpha = 0.7,
                s = 300)
plt.show()
```
